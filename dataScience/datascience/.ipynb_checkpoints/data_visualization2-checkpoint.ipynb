## 层次化索引  ：即是一个轴上面有多个索引

## Series层次化索引

```python
s = pd.Series([1,2,3,4,5,6,7,8,9],index=[['a','a','a','b','b','c','c','d','d'],[1,2,3,1,2,3,1,2,3]])
```

```python
s
```

```python
s.index
```

```python
s['a']
```

```python
s['a':'c']  #对外层索引进行操作
```

```python
s[:,1]
```

```python
s['a',1]
```

```python
s['c',3]
```

```python
#通过unstack方法可以将Series变成一个DataFrame
s.unstack()
```

```python
s.unstack().stack()  #转会层次化索引
```

## 数据分组
    groupby 实现数据分组和分组运算，作用类似于数据透视表

```python
group = df.groupby(df['产地'])   #按照电影的产地进行分组
```

```python
type(group)                      #先定义一个分组变量
```

```python
group.mean()                     #计算分组后各个数据的统计量
```

```python
group.sum()    #求加和
```

```python
df['评分'].groupby(df['年代']).mean()
```

```python
df['年代'] = df['年代'].astype('str')
```

```python
df.groupby(df['产地']).mean()
```

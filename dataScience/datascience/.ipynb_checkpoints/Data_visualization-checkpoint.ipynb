# 数据透视

## 基本形式

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('newdata.xlsx')
```

```python
df
```

```python
pd.set_option('max_columns',100)  #对数据透视表进行设置 最大列为100
pd.set_option('max_rows',200)     #对数据透视表的行进行设置。最大行为200
```

```python
pd.pivot_table(df,index=['上市时间'])  #使用一个索引
```

```python
pd.pivot_table(df,index=['上市时间','城市'])    #想看各个年份和各个城市的上市公司
```

```python
#pd.pivot_table(df,index=['上市时间','城市'],values=['利润'])  来指定需要统计汇总的数据 
```

```python
pd.pivot_table(df,index=['上市时间','城市'],values=['利润'],aggfunc=np.sum)   #统计不同的统计值
```

```python
pd.pivot_table(df,index=['省份'],values=['员工人数','利润'],aggfunc=[np.sum])  
```

```python
table = pd.pivot_table(df,index=['城市'],aggfunc=[np.sum,np.mean],fill_value=0)    #非数值（空值）的内容都比较难以处理，我们可以将其修改为0
```

```python
table = pd.pivot_table(df,index=['城市'],aggfunc=[np.sum,np.mean],fill_value=0,margins=True)   #对于新的统计进行求和
```

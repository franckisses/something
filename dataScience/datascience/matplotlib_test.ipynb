```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'ro',t,t**2,'r',t,t**3,'y')
plt.show()
```

```python
x = np.linspace(-np.pi,np.pi)
y = np.sin(x)
plt.plot(x,y,'r',linewidth=4.0,color='r')
plt.show()
```

```python
line1,line2 = plt.plot(x,y,'r',x,y+1,'g-')
line2.set_antialiased(False)
plt.show()
```

```python
def f(t):
    return t**2
    
t1 = np.arange(0.0,5.0,0.2)
t2 = np.arange(0.0,5.0,0.1)
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(t1,f(t1),'ro')

plt.subplot(212)
plt.plot(t2,np.cos(t2),'g')

plt.show()
```
